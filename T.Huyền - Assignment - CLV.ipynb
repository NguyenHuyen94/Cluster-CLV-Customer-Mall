```python
#TÌM K 
from sklearn.cluster import KMeans
SSE=[]
for cluster in range(1,10):
    kmeans= KMeans (n_clusters= cluster, init = 'k-means++')
    kmeans.fit(df)
    SSE.append(kmeans.inertia_)
print(SSE)
plt.plot(range(1,10), SSE)
plt.show()
#K hợp lý là 3 cụm
```

```python
#tim tam ngau nhien, predict
kmeans = KMeans(n_clusters=3, random_state=0).fit(df)
print('Centers found by scikit-learn:')
kmeansdata= kmeans.cluster_centers_
kmeansdata
```

```python
centroids = pd.DataFrame(kmeansdata, columns = df.columns)
centroids
```

```python
#ve thu centroid vao scatter
centroids = pd.DataFrame(kmeansdata, columns = df.columns)
plt.scatter(df['INCOME'], df['SPEND'], c= 'black')
plt.scatter(centroids['INCOME'],centroids['SPEND'], c = 'red')
plt.xlabel('INCOME')
plt.ylabel('SPEND')
plt.show()
```

```python
import numpy as np
#bat dau apply tung phan tu
diff=1
count=0
while diff != 0:
    i=1
    for indexc, rowc in centroids.iterrows():
        dist_dc =[]
        for indexd, rowd in df.iterrows():
            d1=(rowc['SPEND']-rowd['SPEND'])**2
            d2=(rowc['INCOME']-rowd['INCOME'])**2
            d=np.sqrt(d1+d2)
            dist_dc.append(d)
        df[i]=dist_dc
        i = i+1
    C = []
    for index, row in df.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(3):
            if row[i+1]<min_dist:
                min_dist=row[i+1]
                pos=i+1
        C.append(pos)
    df['Clusters']=C
    Centroids_new= df.groupby('Clusters').mean()[["SPEND","INCOME"]]
    if count == 0:
        diff = 1
        count = count + 1
    else:
        diff = (Centroids_new['SPEND'] - Centroids['SPEND']).sum() + (Centroids_new['INCOME'] - Centroids['INCOME']).sum()
    Centroids = df.groupby('Clusters').mean()[["SPEND","INCOME"]]
Centroids
```

```python
#ve cac diem len do thi
color =['green', 'blue', 'yellow']
#range 3 ở đây là 3 màu, không phải 3 tâm
for k in range (3):
    data=df[df['Clusters'] == k+1]
    plt.scatter(data['INCOME'], data['SPEND'], c = color[k])
plt.scatter(Centroids['INCOME'], Centroids['SPEND'], c = 'black')
plt.xlabel("INCOME")
plt.ylabel("SPEND")
plt.show()
```
